```python
exp_group = {'E-MEXP-1416' : ['g2_g1', 'g4_g3'],
             'E-GEOD-20333' : ['g2_g1'],
             'E-GEOD-7307' : ['g83_g17','g82_g16', 'g72_g15', 'g63_g14', 'g48_g13'],
             'E-GEOD-7621' : ['g1_g2'],
             'E-GEOD-20168' : ['g2_g1']}
```

```python
all_dfs = []
for dfs, exp in zip([df1,df2,df3,df4,df5],exp_group.keys()):
    for group in exp_group[exp]:
        colnames = {f'{group}.p-value' : 'p-value',
                   f'{group}.log2foldchange' : 'log2foldchange'}
        df = dfs[['Gene Name', f'{group}.p-value', f'{group}.log2foldchange']].copy(deep=False)
        df.rename(columns=colnames, inplace=True)
        df['group'] = [group for i in range(len(df))]
        df['experiment'] = [exp for i in range(len(df))]
        all_dfs.append(df)
```
